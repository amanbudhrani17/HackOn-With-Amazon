```python
# Our recommendation based on KEYWORD SCORES.


purchase_frequency = {product['name']: 0 for product in product_data}

user_prompt = "I want to buy a modern t-shirt "
keywords = user_prompt.lower().split()

def calculate_relevance_score(product, keywords, browsing_history, purchase_history):
    score = 0
    for keyword in keywords:
        if (keyword in product['name'].lower() or
            keyword in product['description'].lower()):
            score += 3
    score += (2*browsing_history.count(product) + purchase_history.count(product)) 
    return score + purchase_frequency[product['name']]

def recommend_products(user_prompt, product_data, browsing_history, purchase_history):
    matching_products = []
    for product in product_data:
        score = calculate_relevance_score(product, keywords, browsing_history, purchase_history)
        if score > 0:
            matching_products.append((product, score))
    matching_products.sort(key=lambda x: x[1], reverse=True)
    recommended_products = [product[0] for product in matching_products]
    return recommended_products

browsing_history = [product_data[1], product_data[0]]
purchase_history = [product_data[1], product_data[0], product_data[0]]

for product in recommend_products(user_prompt, product_data, browsing_history, purchase_history):
    print(f"{product['name']} (Score: {calculate_relevance_score(product, keywords, browsing_history, purchase_history)})")
```
